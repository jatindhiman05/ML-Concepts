**1. Setup**

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

**2. Load Data**

```python
df = pd.read_csv("diabetes.csv")
```

```python
df.shape
```

```python
df.columns
```

```python
df.head()
```

** Pima Indians Diabetes Dataset**

| Column Name | Description |
|------------|-------------|
| Pregnancies | Number of times the patient has been pregnant |
| Glucose | Plasma glucose concentration after 2 hours in an oral glucose tolerance test |
| BloodPressure | Diastolic blood pressure (mm Hg) |
| SkinThickness | Thickness of the triceps skin fold (mm) |
| Insulin | 2 hour serum insulin level (mu U/ml) |
| BMI | Body Mass Index (weight in kg / height in m²) |
| DiabetesPedigreeFunction | Likelihood of diabetes based on family history |
| Age | Age of the patient in years |
| Outcome | Target variable indicating diabetes status (1 = Diabetic, 0 = Non Diabetic) |

**4. Split features and target**

```python
X = df.drop(columns=["Outcome"])
y = df["Outcome"]
```

```python
X[:5]
```

```python
y[:5]
```

**5. Train Test Split**

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

```python
print("X shape:", X.shape)
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
```

**6. Scale the features**

Fit scaler only on training data, then transform both training and test data

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)  # fit & transform
X_test_scaled = scaler.transform(X_test)
```

```python
print("Preprocessed training data")
print(X_train_scaled.shape)
print(y_train.shape)

print("-"*30)

print("Preprocessed test data")
print(X_test_scaled.shape)
print(y_test.shape)
```

**7. Model Training**

```python
model = LogisticRegression()
```

```python
# training the model
model.fit(X_train_scaled, y_train)
```

**8. Inference**

```python
# training data inference (predictions)
y_train_pred = model.predict(X_train_scaled)

# test data inference
y_test_pred = model.predict(X_test_scaled)
```

```python
y_train[0:5]
```

```python
y_train_pred[0:5]
```

**Training data performance**

```python
# comparing true labels and predicted labels for training data
train_accuracy = accuracy_score(y_train, y_train_pred)
```

```python
train_accuracy
```

```python
print(f"Training Accuracy {train_accuracy*100:.2f} %")
```

**Test data Performance**

```python
test_accuracy = accuracy_score(y_test, y_test_pred)
```

```python
print(f"Test Accuracy {test_accuracy*100:.2f} %")
```

| Model Behavior | Training Accuracy | Test Accuracy | Meaning |
|---------------|------------------|---------------|---------|
| Underfitting | Low | Low | Model is too simple and fails to learn patterns |
| Overfitting | High | Much lower | Model memorizes training data and does not generalize |
| Best Fit | High | High (similar) | Model learns general patterns and performs well |

**Interpretation**

```python
# not a hard rule

if train_accuracy < 0.8 and test_accuracy < 0.8:
    print("Model is UNDERFITTING")
elif train_accuracy - test_accuracy > 0.1:
    print("Model is OVERFITTING")
else:
    print("Model is performing well")
```
