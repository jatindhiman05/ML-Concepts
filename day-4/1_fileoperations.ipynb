```python
## read a whole file
with open('example.txt','r') as file:
    content = file.read()
    print(content)
```

```python
## read a file line by line
with open('example.txt','r') as file:
    for line in file:
        print(line.strip()) ## strip() remove newline character
```

```python
## writing a file (will override)
with open ('example.txt','w') as file:
    file.write('hello world\n')
    file.write('hey new line\n')
```

```python
## writing a file (without overwriting)
with open ('example.txt','a') as file:
    file.write('here is append operation\n')
```

```python
### Writing a list of lines to a file
lines=['First line \n','Seocnd line \n','Third line\n']
with open('example.txt','a') as file:
    file.writelines(lines)
```

```python
## binary files
# Writing to a binary file
data = b'\x00\x01\x02\xe3\x04'
with open('example.bin', 'wb') as file:
    file.write(data)
```

```python
# Reading a binary file
with open('example.bin', 'rb')as file:
    content = file.read()
    print(content)
```

```python
### Read the content froma source text fiile and write to a destination text file
# Copying a text file
with open('example.txt', 'r') as source_file:
    content = source_file.read()

with open('destination.txt', 'w') as destination_file:
    destination_file.write(content)
```

```python
# The w+ mode in Python is used to open a file for both reading and writing. If the file does not exist, it will be
# created. If the file exists, its content is truncated (i.e., the file is overwritten).
```

```python
### Writing and then reading a file

with open('example.txt','w+') as file:
    file.write("Hello world\n")
    file.write("This is a new line \n")

    ## Move the file cursor to the beginning
    file.seek(0)

    ## Read the content of the file
    content = file.read()
    print(content)
```
