```python
## while loop
## The while loop continues to execute as long as the condition is True.

count = 0

while count<5:
    print(count)
    count = count + 1
```

```python
## loop control statements

## break

for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
## continue
## skips the current iteration and continue with next

for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```

```python
## pass
## the pass statement is a null operation , do nothing
for i in range(5):
    if i == 3:
        print("the number is",i)
        pass
    print(i)
```

```python
## nested loops
# a loop inside a loop

for i in range(3):
    for j in range(3):
        print(f"i: {i}, j : {j}")
```

```python
num = int(input("Enter the no."))

sum = 0
i = 1

while i <= num:
    sum += i
    i = i + 1

print(sum)
```

```python
num = int(input("Enter the no."))

sum = 0
i = 1

for i in range(1,num + 1):
    sum += i

print(sum)
```
