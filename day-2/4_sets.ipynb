```python
## Sets are a built-in data type in Python used to store collections of unique items. They are unordered, meaning
## that the elements do not follow a specific order, and they do not allow duplicate elements. Sets are useful for
## membership tests, eliminating duplicate entries, and performing mathematical set operations like union,
## intersection, difference, and symmetric difference.
```

```python
## create a set
my_set = {1,2,3,4,5}
print(my_set)
print(type(my_set))
```

```python
## create a empty set
my_empty_set = set()
print(type(my_empty_set))
```

```python
my_set = set([1,2,3,3,3,5])
print(type(my_set))
print(my_set)
```

```python
## basic set operation

## adding and removing elements
my_set.add(7)
my_set
```

```python
my_set.remove(2)
my_set
```

```python
my_set.discard(2)
my_set
```

```python
## pop removes random element
removed_elem = my_set.pop()
removed_elem
```

```python
## clear all the elements
my_set.clear()
my_set
```

```python
## set membership test
my_set = {1,2,3,4}
print(type(my_set))
print(3 in my_set)
print(10 in my_set)
```

```python
## mathematical operation
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

### Union
union_set = set1.union(set2)
print(union_set)

### Intersection
intersection_set = set1.intersection(set2)
print(intersection_set)

set1.intersection_update(set2)
print(set1)
```

```python
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

## Difference
print(set1.difference(set2))
```

```python
## symmetric difference
set1.symmetric_difference(set2)
```

```python
## sets methods

set1 = {1,2,3}
set2 = {3,4,5}

print(set1.issubset(set2))
print(set1.issuperset(set2))
```

```python
lst=[1,2,3,4,5,4,2]
set(lst)
```

```python
## counting unique words in text

text = "hi i am jatin dhiman what about you"
words=text.split()

unique_words = set(words)
print(unique_words)
print(len(unique_words))
```
