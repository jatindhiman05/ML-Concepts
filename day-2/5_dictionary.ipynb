```python
## Dictionaries are unordered collections of items. They store data in key-value pairs. Keys must be unique and
## immutable (e.g., strings, numbers, or tuples), while values can be of any type.
```

```python
## creating dictionary
empty_dict = {}
print(type(empty_dict))
```

```python
empty_dict = dict()
print(type(empty_dict))
```

```python
student={"name" : "jatin dhiman","age":20,"grade": "A+"}
```

```python
print(student)
```

```python
## accessing dict elements by key
student["name"]
```

```python
## accessing dict elements using get() method
print(student.get('grade'))
print(student.get('last_name'))
print(student.get('last_name',"Not available"))
```

```python
## Modifying dictionary elements
## dictionary are mutable , so u can add , update or delete elements

print(student)
```

```python
student["age"]=21 ## updated the value for key
print(student)
student["address"] = "India" ## added a new key and val
print(student)
```

```python
del student['grade'] ## delete key and value pair
print(student)
```

```python
## dict methods

keys = student.keys()
print(keys)

values = student.values()
print(values)

items = student.items()
print(items)
```

```python
student_copy = student
print(student)
print(student_copy)
```

```python
student['name']= 'kamal' # this actually changes both the dicts
print(student)
print(student_copy)
```

```python
# we want the update to not happen in the copy created --> shallow copy

student_copy1=student.copy()
print(student_copy1)
```

```python
student['name'] = 'raghav'
```

```python
print(student_copy1)
```

```python
### iterating over dicts
### we can use loops to iterate , keys , values , items

## iterating over keys

for key in student.keys():
    print(key)
```

```python
for value in student.values():
    print(value)
```

```python
for key,value in student.items():
    print(f"{key}: {value}")
```

```python
## Nested Dicts

students = {
    "student1":{"name" : "jatin","age" : 20},
    "student2":{"name" : "kamal","age" : 20}
}

print(students)
```

```python
## access nested dict elems
print(students['student2']['name'])
```

```python
## iterating over nested dicts

for student,student_info in students.items():
    print(f"{student}")
    for key ,value in student_info.items():
        print(f"{key}: {value}")
    print("\n")
```

```python
##  dictionary comprehension

squares = {x: x**2 for x in range(5)}
print(squares)
```

```python
evens = {x:x**2 for x in range(5) if x%2 == 0}
print(evens)
```

```python
lst = [1,2,2,3,3,3,4,4,4,4,5,5]
freq = dict()

for x in lst:
    if x in freq:
        freq[x] += 1
    else:
        freq[x] = 1

freq
```

```python
# merge two dict into one

dict1 = {"a" : 1 , "b":2}
dict2 = {"b" : 3 , "c" : 4}

merged_dict = {**dict1,**dict2}
merged_dict
```
