```python
# What is Seaborn

# Seaborn is a data visualization library built on top of Matplotlib. It provides beautiful and easy to use statistical plots that make complex data easier to
# understand.

# Why Seaborn is Important in Machine Learning

# Seaborn is a high-level visualization library built on top of Matplotlib.
# It makes it easy to create beautiful, statistical graphics for data analysis and model interpretation.

# Why Seaborn is Essential:

# Beautiful by Default: Clean, modern charts with minimal code.
# Statistical Insights: Helps visualize distributions, correlations, and relationships.
# ML Data Exploration: Great for feature analysis, class balance, and outlier detection.
# Integration: Works seamlessly with Pandas DataFrames and NumPy arrays.
# Built on Matplotlib: Same power, much simpler syntax.
# If Matplotlib shows data, Seaborn helps you understand it visually.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Load in-built seaborn dataset

tips = sns.load_dataset("tips")
tips.shape
```

```python
tips.head()
```

```python
# set global styles & color palette
sns.set_style("darkgrid")
sns.set_palette("viridis")
```

```python
## Histogram - distribution of total bill
plt.figure(figsize=(6, 4))
sns.histplot(tips["total_bill"], kde=True, color="teal")
plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

```python
## Box Plot - Detect Outliers
plt.figure(figsize=(6,4))
sns.boxplot(x="day", y="total_bill", data=tips, palette="pastel", hue="day", legend=False)
plt.title("Total Bill by day")
plt.show()
```

```python
## count plot - number of records per day 
plt.figure(figsize=(6,4))
sns.countplot(x="day", data=tips, hue="sex", palette="Set2")
plt.title("Number of Records per day")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="sex", size="size", palette="cool")
plt.title("Bill vs Tip Relationship")
plt.show()
```

```python
# Pair plot - Pair wise Feature relationship
sns.pairplot(tips, hue="sex", diag_kind="kde")
plt.suptitle("Pair-wise reationship")
plt.show()
```

```python
# Heatmap - Feature Correlation
correlation_matrix = tips.corr(numeric_only=True)
plt.figure(figsize=(6,4))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```
